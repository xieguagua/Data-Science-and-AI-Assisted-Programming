4-2Key variable distribution
Conduct frequency analysis and visualization for the re-encoded categorical independent variables (such as educational level and marital status) to confirm the distribution of the sample in key social attributes.

```python
# 1. Distribution of education level
print("--- Education level frequency distribution (%) ---")
education_freq = (df['education'].value_counts(normalize=True) * 100).round(2)
print(education_freq.to_string())

# Draw a bar plot of education level
plt.figure(figsize=(10, 6))
sns.countplot(
    y='education', 
    data=df, 
    order=df['education'].value_counts().index, 
    palette='Pastel1'
)
plt.title('Distribution of respondents\' education levels', fontsize=16)
plt.xlabel('Number of people', fontsize=12)
plt.ylabel('Education level', fontsize=12)
plt.show()


# 2. Distribution of marital status
print("\n--- Marital status frequency distribution (%) ---")
marital_freq = (df['marital_status'].value_counts(normalize=True) * 100).round(2)
print(marital_freq.to_string())

# Draw a pie chart of marital status
plt.figure(figsize=(8, 8))
# Filter out categories with too few labels and use autopct to display percentages
plt.pie(
    marital_freq.values, 
    labels=marital_freq.index, 
    autopct='%1.1f%%', 
    startangle=90, 
    colors=sns.color_palette("Set2")
)
plt.title('Distribution of respondents\' marital status', fontsize=16)
plt.show()
```

Logarithmic income (log-income) is a continuous variable that has undergone logarithmic transformation. We use descriptive statistics and histograms to examine its central tendency and dispersion.

```python
# 1. Descriptive statistics of logarithmic income
print("--- Descriptive statistics of logarithmic income (income) ---")
print(df['income'].describe().to_string())

# 2. Draw a histogram
plt.figure(figsize=(10, 6))
sns.histplot(
    df['income'], 
    bins=30, 
    kde=True, 
    color='#C20E1A', 
    edgecolor='black'
)
plt.title('Distribution of respondents\' logarithmic income', fontsize=16)
plt.xlabel('Logarithmic income (Log-Income)', fontsize=12)
plt.ylabel('Frequency/Density', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
