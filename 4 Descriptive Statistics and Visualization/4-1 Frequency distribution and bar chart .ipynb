4.1 Analysis of Happiness Distribution (Dependent Variable) 
Happiness is the core dependent variable in this analysis. We examined its frequency distribution to understand the proportion of different levels of happiness within the sample.(The levels are typically 1 = extremely unhappy, 5 = extremely happy)

```python
# 1. Calculate the frequency and relative frequency of happiness
happiness_counts = df['happiness'].value_counts().sort_index()
happiness_freq = (df['happiness'].value_counts(normalize=True).sort_index() * 100).round(2)

print("--- Happiness frequency distribution ---")
print(happiness_counts.to_string())
print("\n--- Happiness relative frequency distribution (%) ---")
print(happiness_freq.to_string())


# 2. Draw a bar plot of happiness
plt.figure(figsize=(9, 6))
# Use sns.barplot() to draw a bar plot
sns.barplot(x=happiness_counts.index, y=happiness_counts.values, color='#4CAF50')

# Set chart title and labels
plt.title('Distribution of respondents\' happiness levels', fontsize=16)
plt.xlabel('Happiness level (1-Very unhappy, 5-Very happy)', fontsize=12)
plt.ylabel('Number of people', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
