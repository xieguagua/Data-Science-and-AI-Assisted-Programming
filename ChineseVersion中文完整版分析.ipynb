1.数据科学与人工智能辅助编程介绍


第一部分、什么是数据科学？

在传统的文科语境下，我们强调“文本细读”或“理论思辨”。而在数据科学语境下，我们强调“从海量杂乱数据中提炼规律”。

数据科学并不是只有高深的算法，它主要由以下四个核心环节组成：

1. 数据获取与读取 

传统痛点：CGSS 数据通常是 Stata (.dta) 格式，用 Excel 打不开，用 SPSS 打开又很慢。

通过编程，我们可以瞬间读取包含几万行、几百列的数据文件，并保留所有的元数据（比如变量标签“A36”代表“幸福感”）。

2. 数据清洗—— 这是最耗时的一步

场景（CGSS特有）：

CGSS 中，“不知道”可能编码为 98，“拒绝回答”是 99，“不适用”是 -8。

如果不处理直接算平均值，99分会让你的结果严重失真。

数据科学视角：我们需要编写一套逻辑，告诉计算机：“如果变量A的值在 [98, 99, -8] 列表中，请将其视为‘空值’，不要参与计算。”

优势：这叫“可复现性”。可以保留论文工作痕迹。

3. 探索性分析

场景：你想知道“收入”和“幸福感”有没有关系。

数据科学视角：不直接跑回归，而是先进行基本描述性分析，再画图（散点图、热力图），通过可视化手段，直观地看到数据的分布形状，发现异常点。

4. 统计建模 

场景：验证假设。

数据科学视角：从简单的 OLS 回归到复杂的机器学习模型。

第二部分：什么是 AI 辅助编程？

AI 辅助编程（基于 LLM 的代码生成技术） 的出现，并不是简单的“自动写代码”，而是实现了一种新的人机协作研究范式：

1. 核心原理：基于语义的逻辑映射

传统编程要求研究者必须掌握计算机的句法规则才能与机器沟通。而 AI 辅助编程允许研究者使用自然语言 描述研究意图，由大模型在后台完成从“研究意图”到“执行代码”的映射。

传统模式：

研究者思维：“我要清洗数据。”

认知壁垒：必须记忆 df.loc[df['income'] < 0, 'income'] = np.nan。

AI 辅助模式：

研究者思维：“将收入变量中的负值编码为缺失值。”

抽象层提升：你只负责定义逻辑规则，AI 负责处理实现细节。

学术启示：这标志着编程能力的门槛，从“记忆代码”转变为“清晰定义研究问题”。

2. AI 在定量研究中的三个功能维度

在处理如 CGSS 2023 这样复杂的微观调查数据时，AI 插件实际上承担了以下三个学术角色：

角色 A：概念的操作化执行者

社会学研究的核心是将抽象概念转化为可测量的指标。AI 负责将你的操作化定义迅速转化为代码。

角色 B：交互式错误诊断 

编程报错本质上是逻辑的不自洽或环境的不匹配。AI 此刻充当的是技术审查员。

角色 C：方法论的最佳实践顾问

这是 AI 最具潜力的高阶功能。它基于海量的代码库和统计文档，能提供符合主流学术规范的建议。

3.2 数据读入 **记得更改文件路径**

```python
file_path = r"C:\Users\EddieYeung\Desktop\谢谢的作业4\cgss2023.dta"
df_raw = pd.read_stata(file_path, convert_categoricals=False)
```

3.3 变量选择

```python
vars_map = {
    'a36': 'happiness',      # 因变量
    'a8a': 'income',         # 自变量
    'a15': 'health',
    'a7a': 'education',
    'a69': 'marital_status',
    'a43a': 'social_class',
    'a25a': 'residence',
    'a285': 'internet_freq',
    'a2': 'gender',
    'a10': 'politics',
    'a3a': 'birth_year',
}
df = df_raw[vars_map.keys()].rename(columns=vars_map).copy()
```

3.4 数据清洗

```python
#3.4.1 小于0全部标记缺失
for col in df.columns:
    # 仅对数值型列做小于0判断，避免字符串与整数比较报错
    if pd.api.types.is_numeric_dtype(df[col]):
        df.loc[df[col] < 0, col] = pd.NA
#3.4.2 特殊规则处理
df.loc[df['happiness'] > 5, 'happiness'] = pd.NA
df.loc[df['income'] >= 9999996, 'income'] = pd.NA
df.loc[df['education'] == 14, 'education'] = pd.NA
df['age'] = 2023 - df['birth_year']
#3.4.3 变量再编码
# 教育程度
edu_map = {
    1: "小学及以下", 2: "小学及以下", 3: "小学及以下", 4: "初中",
    5: "高中_中专", 6: "高中_中专", 7: "高中_中专", 8: "高中_中专",
    9: "大专", 10: "大专", 11: "大学本科", 12: "大学本科", 13: "研究生及以上"
}
marry_map = {1: '未婚', 2: '未婚', 3: '已婚', 4: '已婚', 5: '已婚', 6: '离婚', 7: '丧偶'}
res_map = {1: '城市_城乡', 2: '城市_城乡', 3: '乡镇', 4: '乡镇'}
pol_map = {1: '非党员', 2: '非党员', 3: '非党员', 4: '共产党员'}

df['education'] = df['education'].map(edu_map)
df['marital_status'] = df['marital_status'].map(marry_map)
df['residence'] = df['residence'].map(res_map)
df['politics'] = df['politics'].map(pol_map)

# 收入取对数
df['income'] = np.log1p(df['income'].astype(float))
#3.4.4 删除所有包含缺失值的行
df.dropna(inplace=True)

```

5 双变量和多变量分析、表格输出与可视化

```python
# 1. 数据清洗与筛选
# -------------------------------------------------------
# 定义参与分析的变量列表
analysis_cols = [
    'happiness', 'income', 'health', 'education', 
    'marital_status', 'social_class', 'residence', 
    'internet_freq', 'gender', 'politics', 'age'
]

# 去除缺失值 (Listwise Deletion)
model_df = df[analysis_cols].dropna().copy()

# 收入取对数 (处理偏态分布，避免数值过大影响回归)
# 使用 log1p (log(x+1)) 避免 0 值报错
model_df['ln_income'] = np.log1p(model_df['income'])

print(f"进入回归模型的有效样本量: {len(model_df)}")
```

5.1 双变量相关性分析 (Correlation Matrix)主要关注数值型和定序型变量之间的皮尔逊相关系数。
目的：初步判断变量间关系，并确保自变量间相关系数不过高（通常需 $|r| < 0.7$），以避免多重共线性。

```python
# 筛选数值/定序变量
corr_vars = ['happiness', 'ln_income', 'health', 
             'social_class', 'internet_freq', 'age']

# 计算相关矩阵
corr_matrix = model_df[corr_vars].corr()

# 绘制热力图
plt.figure(figsize=(10, 8))
mask = np.triu(np.ones_like(corr_matrix, dtype=bool)) # 只显示下三角

sns.heatmap(corr_matrix, 
            mask=mask, 
            annot=True, 
            fmt=".2f", 
            cmap='RdBu_r', 
            center=0,
            square=True, 
            cbar_kws={"shrink": .8})

plt.title('核心变量相关性热力图', fontsize=15)
plt.show()
```

5.2 多变量层级回归分析 (Hierarchical Regression)
采用 OLS (最小二乘法) 构建三个嵌套模型。

```python
# 定义回归公式
# -------------------------------------------------------
# 模型 1: 控制变量 (性别、年龄、婚姻、户口、政治面貌)
formula_m1 = 'happiness ~ C(gender) + age + C(marital_status) + C(residence) + C(politics)'

# 模型 2: + 社会经济地位 (收入对数、教育、社会阶层)
formula_m2 = formula_m1 + ' + ln_income + education + social_class'

# 模型 3: + 生活质量与技术 (健康、互联网频率)
formula_m3 = formula_m2 + ' + health + internet_freq'

# 拟合模型
# -------------------------------------------------------
m1 = smf.ols(formula_m1, data=model_df).fit()
m2 = smf.ols(formula_m2, data=model_df).fit()
m3 = smf.ols(formula_m3, data=model_df).fit()

# 输出结果提示
print("模型拟合完成。")
print(f"Model 3 R-squared: {m3.rsquared:.3f}")
```

5.3 回归结果汇总表 (Regression Table)汇总输出三个模型的系数、标准误与显著性星号

```python
# 生成学术规范表格
results_table = summary_col(
    [m1, m2, m3], 
    stars=True, 
    float_format='%0.3f',
    model_names=['Model 1\n(Demo)', 'Model 2\n(+SES)', 'Model 3\n(+Full)'],
    info_dict={
        'N': lambda x: "{0:d}".format(int(x.nobs)),
        'R2': lambda x: "{:.3f}".format(x.rsquared),
        'Adj. R2': lambda x: "{:.3f}".format(x.rsquared_adj)
    },
    regressor_order=['Intercept', 'age', 'ln_income', 'education', 'social_class', 'health'] # 可选：指定显示顺序
)

print(results_table)
```

5.4 结果可视化：系数森林图 (Coefficient Plot)
可视化全模型（Model 3）中各变量的标准化效应或原始效应。

解读：误差棒（置信区间）若不跨越 0 刻度线，表示该变量对幸福感有显著影响。

```python
params = m3.params.drop('Intercept', errors='ignore') # 去除截距项
conf = m3.conf_int().drop('Intercept', errors='ignore')
conf.columns = ['Lower', 'Upper']

coef_df = pd.DataFrame({
    'coef': params,
    'err_low': params - conf['Lower'],
    'err_high': conf['Upper'] - params
})

main_vars = ['age', 'ln_income', 'education', 'social_class', 'health', 'internet_freq']
plot_df = coef_df.loc[coef_df.index.intersection(main_vars)]

plt.figure(figsize=(10, 6))
plt.errorbar(
    plot_df['coef'], 
    plot_df.index, 
    xerr=[plot_df['err_low'], plot_df['err_high']], 
    fmt='o', 
    color='black', 
    ecolor='gray', 
    capsize=5,
    label='95% CI'
)

plt.axvline(x=0, color='red', linestyle='--', linewidth=1)
plt.title('主要变量回归系数图 (Coefficient Plot)', fontsize=14)
plt.xlabel('回归系数 (Coefficient)')
plt.grid(axis='x', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

```python
summary = pd.DataFrame({
    'Coef': m3.params,
    'P-value': m3.pvalues,
    'Conf_Lower': m3.conf_int()[0],
    'Conf_Upper': m3.conf_int()[1]
})

def get_significance(p):
    if p < 0.01: return '*** (极显著)'
    elif p < 0.05: return '** (显著)'
    elif p < 0.1: return '* (边缘显著)'
    else: return '不显著'

summary['Significance'] = summary['P-value'].apply(get_significance)

focus_vars = ['ln_income', 'social_class', 'health', 'education', 'internet_freq']
print("--- 核心假设验证结果 ---")
for var in focus_vars:
    if var in summary.index:
        row = summary.loc[var]
        effect = "正向" if row['Coef'] > 0 else "负向"
        print(f"变量 [{var}]: 效应为{effect} ({row['Coef']:.3f}), 统计上 {row['Significance']}")
```

5.5 实证结果深度解读 
基于模型 3 的回归表格与系数可视化图，我们对“幸福感的影响机制”得出以下核心社会学发现：
1.“伊斯特林悖论”的微观验证：绝对收入的失灵这是一个极具社会学意义的发现：收入（ln_income） 在控制了其他变量后，对幸福感的影响不显著（$p > 0.1$，置信区间跨越 0 线）。与此同时，社会阶层（social_class） 呈现显著的正向影响（$p < 0.001$）。解读：这表明在当前的社会语境下，“觉得自己处于什么位置”比“实际赚多少钱”更能决定幸福感。相对剥夺感（Relative Deprivation）理论在此得到了验证——人们的痛苦往往不源于贫穷本身，而源于与周围人的比较以及对自身地位的低评价。
2.身体是革命的本钱：健康的主导性从系数图可见，健康（health） 的标准化效应最强，且显著性极高。这再次确认了身心健康是幸福感的基石。无论个体处于何种社会阶层，健康状况的恶化都会对幸福感产生“一票否决”式的打击。
3.技术祛魅：互联网的非显著效应互联网使用频率（internet_freq） 的回归系数为负且统计上不显著。这反驳了简单的“技术赋能”论调。高频的互联网使用并未直接转化为更高的生活满意度，反而可能因“信息过载”或“网络社会比较”带来潜在的负面情绪（虽然本模型中未达显著水平，但系数方向值得深思）。
4.教育的隐性红利教育（education）在模型中保持了边缘显著的正向影响。这说明受教育程度除了通过转化收入来间接影响幸福外，还能通过提升个体的认知能力、社会适应力直接提升主观福祉。

6 线性回归模型分析、多变量、表格与可视化

6.1 构建分层回归的三个模型：基准模型（空模型）、核心模型、交互效应模型

分层回归（Hierarchical Regression）的核心逻辑是通过分步骤加入变量，来观察每一组新变量到底能为解释因变量（幸福感）带来多大的额外贡献。模型 1：基准模型仅放入人口学控制变量（如：性别、年龄、婚姻、户口、政治面貌）。作用是排除干扰，确立一个比较的基准线。在控制了人口学背景后，模型二核心模型加入我们真正试图解释的研究变量。如果 R^2 出现大幅提升，就证明了核心变量才是影响幸福感的关键变量an'aib。模型三交互模型探索变量之间是否存在更复杂的关系（调节效应）。比如：收入对幸福感的影响，会不会因为健康状况的不同而不同？

```python
# 模型 1：基准模型 (人口学控制变量)
# happiness ~ 性别 + 婚姻 + 政治 + 户口 + 年龄
formula1 = 'happiness ~ gender + marital_status + politics + residence + age'
model1 = smf.ols(formula1, data=df).fit()

# 模型 2：核心模型 (加入 SES 和 互联网使用)
# happiness ~ (模型1变量) + 收入 + 健康 + 教育 + 阶层 + 互联网
formula2 = 'happiness ~ gender + marital_status + politics + residence + age + \
            income + health + education + social_class + internet_freq'
model2 = smf.ols(formula2, data=df).fit()

# 模型 3：交互效应模型 (探索 收入*健康)
# 看看 "收入" 对幸福感的影响，是否会因为 "健康" 状况不同而变化？
formula3 = 'happiness ~ gender + marital_status + politics + residence + age + \
            education + social_class + internet_freq + income * health'
model3 = smf.ols(formula3, data=df).fit()

print("三个模型运行完毕！")
print(f"模型 1 (基准) R²: {model1.rsquared:.3f}")
print(f"模型 2 (核心) R²: {model2.rsquared:.3f} (加入核心变量后的提升)")
print(f"模型 3 (交互) R²: {model3.rsquared:.3f}")
```

6.2 回归结果对比

```python
def get_model_results(model, model_name):
    """提取模型的系数和星号，做成漂亮的列"""
    # 提取系数和P值
    params = model.params
    pvalues = model.pvalues
    
    # 格式化函数：系数 + 星号
    def format_func(idx):
        coef = params[idx]
        pval = pvalues[idx]
        stars = ''
        if pval < 0.001: stars = '***'
        elif pval < 0.01: stars = '**'
        elif pval < 0.05: stars = '*'
        return f"{coef:.3f}{stars}"
    
    return pd.Series([format_func(i) for i in params.index], index=params.index, name=model_name)

# 生成三个模型的列
col1 = get_model_results(model1, "模型1 (基准)")
col2 = get_model_results(model2, "模型2 (核心)")
col3 = get_model_results(model3, "模型3 (交互)")

# 合并成一张大表
table = pd.concat([col1, col2, col3], axis=1)

# 把 R-squared 和 样本量 N 加在底部
table.loc['R-squared'] = [f"{model1.rsquared:.3f}", f"{model2.rsquared:.3f}", f"{model3.rsquared:.3f}"]
table.loc['Observations'] = [int(model1.nobs), int(model2.nobs), int(model3.nobs)]

# 填充空值 (有些变量在模型1里没有)
table = table.fillna('')

print("\n--- 回归分析结果汇总表 ---")
print(table)
```

6.3 可视化回归系数-基于模型2

```python
# 提取模型2的数据 (去掉截距项 Intercept)
plot_data = pd.DataFrame({
    'coef': model2.params.drop('Intercept'),
    'pvalue': model2.pvalues.drop('Intercept'),
    'conf_lower': model2.conf_int()[0].drop('Intercept'),
    'conf_upper': model2.conf_int()[1].drop('Intercept')
})

# 按系数大小排序，画图更好看
plot_data = plot_data.sort_values('coef')

# 设置颜色：显著(P<0.05)用红色，不显著用灰色
plot_data['color'] = plot_data['pvalue'].apply(lambda x: '#D62728' if x < 0.05 else '#7F7F7F')

# 画图
plt.figure(figsize=(12, 8))

# 画水平条形图 (带有误差棒/置信区间)
# 误差棒长度 = (上界 - 下界) / 2
errors = (plot_data['conf_upper'] - plot_data['conf_lower']) / 2
plt.barh(plot_data.index, plot_data['coef'], xerr=errors, color=plot_data['color'], capsize=4, alpha=0.8)

# 添加装饰
plt.axvline(x=0, color='black', linestyle='--', linewidth=1) # 0刻度线
plt.title('影响幸福感的因素分析 (OLS回归系数)', fontsize=16, fontweight='bold')
plt.xlabel('回归系数 (Coefficient) \n (正数表示提升幸福感，负数表示降低)', fontsize=12)
plt.ylabel('自变量', fontsize=12)
plt.grid(axis='x', linestyle='--', alpha=0.5)

# 添加图例说明
from matplotlib.patches import Patch
legend_elements = [Patch(facecolor='#D62728', label='显著 (P < 0.05)'),
                   Patch(facecolor='#7F7F7F', label='不显著 (P >= 0.05)')]
plt.legend(handles=legend_elements, loc='lower right')

plt.tight_layout()
plt.show()
```

7 广义回归模型分析

```python
# 因变量a36是1-5的评分，通常为高斯分布(Gaussian)或使用有序回归。
# 这里使用标准的GLM (默认Family为Gaussian)
formula = """
happiness ~ income + health + C(education, Treatment('小学及以下')) + 
            C(marital_status, Treatment('未婚')) + social_class + 
            C(residence) + internet_freq + gender + C(politics)
"""
model = smf.glm(formula=formula, data=df, family=sm.families.Gaussian()).fit()

print(model.summary())
```

1.健康与社会阶层是幸福感最稳健的支柱。
自评健康（β=0.139, p<0.001） 与 主观社会阶层（β=0.100, p<0.001） 具有极强的统计显著性，是提升幸福感最可靠的因素。
2.教育存在“门槛效应”而非简单线性关系。
与参照组“小学及以下”相比，高中/中专（β=0.103, p=0.002）、大专（β=0.126, p=0.002）和本科（β=0.122, p=0.003） 学历者的幸福感显著更高，且效应值相近。初中（p=0.358）和研究生及以上（p=0.929） 学历者的幸福感与小学组无统计学差异。这表明中等及以上职业教育与高等教育是提升幸福感的关键教育“门槛”。
3.婚姻状况影响呈现两极分化。
离婚状态带来显著的幸福感损伤（β=-0.191, p<0.001），是所有变量中负向效应最强的。已婚状态显著提升幸福感（β=0.094, p=0.002）。
丧偶状态的影响在统计上边缘显著（β=0.080, p=0.074），未达到常规显著性水平（p<0.05）。
4.政治面貌与互联网使用具有显著的负向关联。
非党员身份（β=-0.120, p<0.001） 与 更高的互联网使用频率（β=-0.016, p=0.037） 均与较低的幸福感显著相关。后者提示需要关注网络使用的潜在负面影响（如信息过载、社会比较、网络成瘾）。
5.多个传统变量影响不显著。
个人收入（p=0.652）、性别（p=0.154）、城乡居住地（p=0.188） 在控制其他变量后，对幸福感的直接净效应均无统计学显著性。

8 决策树与机器学习

8.1 特征工程：虚拟变量转换

目的： 机器学习模型（尤其是决策树）无法直接处理文本标签或分类变量（如“大学本科”）。此步骤将分类变量转换为数值型虚拟变量（One-Hot Encoding），以便模型能够进行数值计算。

```python
from sklearn.model_selection import train_test_split

# 识别所有非数值型（object）的分类列
categorical_cols = df.select_dtypes(include=['object']).columns

# 使用pd.get_dummies进行独热编码
# drop_first=True可以避免多重共线性
df_processed = pd.get_dummies(df, columns=categorical_cols, drop_first=True)

# 打印新数据集的前几行和维度，作为检查
print("【虚拟变量转换后】数据集的前5行:")
print(df_processed.head().to_string())
print(f"新维度: {df_processed.shape}")
```

8.2 数据划分：训练集与测试集

目的： 这是机器学习的标准流程。我们将数据集随机划分为训练集（Training Set）和测试集（Test Set）。模型只在训练集上学习规律，然后在测试集上进行预测，以评估模型的泛化能力（即在未知数据上的表现）。

```python
# 目标变量 (Y)
Y = df_processed['happiness']

# 特征变量 (X)：排除目标变量
# axis=1 表示删除列
X = df_processed.drop('happiness', axis=1) 

# 将数据划分为训练集和测试集 (80% 训练 / 20% 测试)
# random_state=42 保证结果可复现
X_train, X_test, Y_train, Y_test = train_test_split(
    X, Y, 
    test_size=0.2, 
    random_state=42
)

print(f"训练集样本量: {len(X_train)}")
print(f"测试集样本量: {len(X_test)}")
```

8.3 决策树回归模型拟合与评估

目的： 使用 sklearn 的 DecisionTreeRegressor 拟合模型，并使用 R-squared (R²) 和 均方误差 (MSE) 在测试集上评估模型的预测效果。

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_squared_error, r2_score

# 1. 初始化模型
# max_depth=5 限制树的深度，防止模型过度拟合（overfitting）训练数据
dt_model = DecisionTreeRegressor(random_state=42, max_depth=5) 

# 2. 训练模型（拟合）
dt_model.fit(X_train, Y_train)

# 3. 在测试集上进行预测
Y_pred = dt_model.predict(X_test)

# 4. 评估模型性能
mse = mean_squared_error(Y_test, Y_pred)
r2 = r2_score(Y_test, Y_pred)

print("="*40)
print("     决策树回归模型性能评估")
print("="*40)
print(f"均方误差 (MSE): {mse:.4f}")
print(f"R-squared (R2): {r2:.4f}")
```

8.4 决策树：特征重要性可视化

目的： 提取模型计算出的每个特征的重要性得分。这个分数代表了该特征在构建决策树时，对降低预测误差的平均贡献。它是判断变量影响力的重要指标。

```python
# 1. 获取特征重要性
feature_importances = dt_model.feature_importances_

# 2. 转换为 Series，使用 X_train 的列名作为索引，这样就不会有 Key Error
importance_series = pd.Series(feature_importances, index=X_train.columns)

# 3. 转换为 DataFrame，方便排序和绘图
importance_df = importance_series.sort_values(ascending=False).reset_index()
importance_df.columns = ['Feature', 'Importance']

# 4. 打印最重要的特征
print("\n--- 决策树特征重要性 (Top 10) ---")
print(importance_df.head(10).to_string())

# 5. 可视化特征重要性 
plt.figure(figsize=(10, 6))
sns.barplot(x='Importance', y='Feature', data=importance_df.head(10), palette='viridis')
plt.title('决策树模型特征重要性排名 (Top 10)', fontsize=16)
plt.xlabel('重要性得分', fontsize=12)
plt.ylabel('特征变量', fontsize=12)
plt.tight_layout()
plt.show()
```

决策树模型会根据每个特征（自变量）在分割节点时带来的信息增益或误差减少，来计算其重要性得分。得分越高，该特征在预测幸福感方面的贡献越大。无论是决策树模型还是之前的 GLM 模型，都确认了主观感受类变量（health 和 social_class）是预测幸福感的最稳健支柱，而非纯粹的客观经济指标。

8.5 决策树结构图可视化 (使用 Matplotlib)

决策树结构图是模型决策过程的白箱展示。它将复杂的预测逻辑分解为一系列简单的 “是/否” 问题，从根节点（Root Node）开始，逐步拆分样本，直至找到具有相似幸福感得分的子集。

目的： 将决策树的结构以流程图的形式展示出来，让学习者直观地理解模型是如何根据特征的阈值进行分层决策的。

我们使用 `sklearn.tree.plot_tree` 结合 `matplotlib` 来绘制决策树结构，以直观展示模型的决策路径。

**注意：** 为了保持图表在 Notebook 中的可读性，我们仍然仅展示树的**前四层** (`max_depth=4`)，并且限制了字体大小。

```python
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt

# 清理任何之前的图表实例，确保只绘制当前图形
plt.clf()
plt.close('all')

# 调整图表大小以适应决策树结构
# 调大尺寸以容纳更多文本和节点
plt.figure(figsize=(25, 18)) 

# 使用 plot_tree 函数绘制决策树
plot_tree(
    dt_model,
    feature_names=X_train.columns.tolist(), # 特征名
    class_names=['1', '2', '3', '4', '5'], # 幸福感等级
    filled=True,
    rounded=True,
    fontsize=9,
    max_depth=4,  # 尝试展示前4层
    proportion=True # 显示每个节点的样本比例而非数量，图表信息更丰富
)

plt.title('决策树回归模型结构图 (前4层)', fontsize=18)
# 强制显示图形
plt.show()
```

决策路径解读（根节点至第二层）：

根节点（Root Node）： 树的起点通常是最重要的特征。你将看到第一个分割条件是 health <= [某个阈值]。

第一层分割：

左侧路径（health 较差）： 如果受访者自评健康低于阈值，样本将被导向左侧。这些样本的平均幸福感得分（value）会较低。

右侧路径（health 较好）： 如果受访者自评健康高于阈值，样本将被导向右侧。这些样本的平均幸福感得分（value）会较高。

第二层分割： 在健康状况已经划分的基础上，模型通常会引入第二重要的特征进行更精细的拆分，例如 social_class 或 income。这展示了模型如何结合多个因素，逐步将样本分配到不同的预测得分区间。

核心意义： 这个图直观地说明了决策树模型如何将“健康”作为首要判断标准，然后分层利用其他因素（如社会阶层或对数收入）来进行细致校正，最终得出对个体幸福感的预测。
