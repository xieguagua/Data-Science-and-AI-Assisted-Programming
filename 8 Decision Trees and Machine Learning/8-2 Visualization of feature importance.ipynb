Objective: Extract the importance score for each feature calculated by the model. This score represents the average contribution of the feature to reducing the prediction error when constructing the decision tree. It is an important indicator for assessing the influence of the variable.

The decision tree model calculates the importance score of each feature (independent variable) based on the information gain or error reduction it brings when splitting the nodes. The higher the score, the greater the contribution of that feature to predicting happiness. 8.5 Visualization of Decision Tree Structure Diagram (Using Matplotlib) 
The decision tree structure diagram is a transparent display of the model's decision-making process. It breaks down the complex prediction logic into a series of simple "yes/no" questions, starting from the root node (Root Node), and gradually splits the samples until a subset with similar happiness scores is found. 
Objective: To present the structure of the decision tree in the form of a flowchart, enabling learners to intuitively understand how the model makes hierarchical decisions based on the thresholds of the features. 
We use `sklearn.tree.plot_tree` in conjunction with `matplotlib` to draw the decision tree structure, thereby visually presenting the decision-making path of the model. 
**Note:** To ensure the readability of the chart within the Notebook, we are still only displaying the first four levels of the tree (`max_depth=4`), and have limited the font size.

```python
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt

# Clear any previous chart instances to ensure only the current graph is drawn
plt.clf()
plt.close('all')

# Adjust chart size to accommodate the decision tree structure
# Increase size to hold more text and nodes
plt.figure(figsize=(25, 18)) 

# Use plot_tree function to draw the decision tree
plot_tree(
    dt_model,
    feature_names=X_train.columns.tolist(), # Feature names
    class_names=['1', '2', '3', '4', '5'], # Happiness levels
    filled=True,
    rounded=True,
    fontsize=9,
    max_depth=4,  # Try to display the first 4 layers
    proportion=True # Display the proportion of samples in each node rather than counts, making the chart more informative
)

plt.title('Decision Tree Regression Model Structure Diagram (First 4 Layers)', fontsize=18)
# Force display the graph
plt.show()
```

Decision path interpretation (from root node to second level): 
Root Node: The starting point of the tree is usually the most important feature. You will see that the first splitting condition is health <= [a certain threshold]. 
First layer division: 
Left path (poor health): If the respondents self-assess their health as being below the threshold, the samples will be directed to the left. The average happiness score (value) of these samples will be lower. 
Right path (healthier): If the respondents rate their own health above the threshold, the samples will be directed to the right. The average happiness score (value) of these samples will be higher. 
Second layer segmentation: After the health status has been classified, the model typically introduces the second most important feature for a more detailed division, such as social_class or income. This demonstrates how the model combines multiple factors and gradually assigns samples to different prediction score ranges. 
Core meaning: This graph visually demonstrates how the decision tree model takes "health" as the primary criterion for judgment, and then uses other factors (such as social class or logarithmic income) in a hierarchical manner for detailed correction, ultimately arriving at a prediction of individual well-being.
