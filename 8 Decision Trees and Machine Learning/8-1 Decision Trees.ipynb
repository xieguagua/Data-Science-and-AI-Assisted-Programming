8.1 Feature Engineering: Transformation of Dummy Variables
Objective: Machine learning models (especially decision trees) cannot directly handle text labels or categorical variables (such as "Bachelor's degree"). This step converts categorical variables into numerical dummy variables (One-Hot Encoding) so that the models can perform numerical calculations.

```python
from sklearn.model_selection import train_test_split

# Identify all non-numeric (object) categorical columns
categorical_cols = df.select_dtypes(include=['object']).columns

# Use pd.get_dummies for one-hot encoding
# drop_first=True can avoid multicollinearity
df_processed = pd.get_dummies(df, columns=categorical_cols, drop_first=True)

# Print the first few rows and dimensions of the new dataset as a check
print("【After dummy variable conversion】First 5 rows of the dataset:")
print(df_processed.head().to_string())
print(f"New dimensions: {df_processed.shape}")
```

8.2 Data Division: Training Set and Test Set
Objective: This is the standard process of machine learning. We randomly divide the dataset into a training set and a test set. The model learns patterns only on the training set and then makes predictions on the test set to evaluate the model's generalization ability (i.e., its performance on unknown data).

```python
# Target variable (Y)
Y = df_processed['happiness']

# Feature variables (X): exclude the target variable
# axis=1 means delete columns
X = df_processed.drop('happiness', axis=1) 

# Split the data into training and test sets (80% training / 20% testing)
# random_state=42 ensures reproducible results
X_train, X_test, Y_train, Y_test = train_test_split(
    X, Y, 
    test_size=0.2, 
    random_state=42
)

print(f"Training set sample size: {len(X_train)}")
print(f"Test set sample size: {len(X_test)}")
```

8.3 Decision Tree Regression Model Fitting and Evaluation
Objective: Fit a model using sklearn's DecisionTreeRegressor, and evaluate the predictive performance of the model on the test set using R-squared (R²) and Mean Squared Error (MSE).

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_squared_error, r2_score

# 1. Initialize the model
# max_depth=5 limits the depth of the tree to prevent overfitting to the training data
dt_model = DecisionTreeRegressor(random_state=42, max_depth=5) 

# 2. Train the model (fit)
dt_model.fit(X_train, Y_train)

# 3. Make predictions on the test set
Y_pred = dt_model.predict(X_test)

# 4. Evaluate model performance
mse = mean_squared_error(Y_test, Y_pred)
r2 = r2_score(Y_test, Y_pred)

print("="*40)
print("     Decision Tree Regression Model Performance Evaluation")
print("="*40)
print(f"Mean Squared Error (MSE): {mse:.4f}")
print(f"R-squared (R2): {r2:.4f}")
```
