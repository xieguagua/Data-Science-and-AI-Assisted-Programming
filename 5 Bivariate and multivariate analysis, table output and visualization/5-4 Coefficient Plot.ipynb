The standardized effects or original effects of each variable in the visualized full model (Model 3). 
Interpretation: If the error bars (confidence intervals) do not cross the 0 axis, it indicates that this variable has a significant impact on happiness.

```python
params = m3.params.drop('Intercept', errors='ignore') # Remove intercept term
conf = m3.conf_int().drop('Intercept', errors='ignore')
conf.columns = ['Lower', 'Upper']

coef_df = pd.DataFrame({
    'coef': params,
    'err_low': params - conf['Lower'],
    'err_high': conf['Upper'] - params
})

main_vars = ['age', 'ln_income', 'education', 'social_class', 'health', 'internet_freq']
plot_df = coef_df.loc[coef_df.index.intersection(main_vars)]

plt.figure(figsize=(10, 6))
plt.errorbar(
    plot_df['coef'], 
    plot_df.index, 
    xerr=[plot_df['err_low'], plot_df['err_high']], 
    fmt='o', 
    color='black', 
    ecolor='gray', 
    capsize=5,
    label='95% CI'
)

plt.axvline(x=0, color='red', linestyle='--', linewidth=1)
plt.title('Coefficient plot of main variables', fontsize=14)
plt.xlabel('Coefficient')
plt.grid(axis='x', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

```python
summary = pd.DataFrame({
    'Coef': m3.params,
    'P-value': m3.pvalues,
    'Conf_Lower': m3.conf_int()[0],
    'Conf_Upper': m3.conf_int()[1]
})

def get_significance(p):
    if p < 0.01: return '*** (Highly significant)'
    elif p < 0.05: return '** (Significant)'
    elif p < 0.1: return '* (Marginally significant)'
    else: return 'Not significant'

summary['Significance'] = summary['P-value'].apply(get_significance)

focus_vars = ['ln_income', 'social_class', 'health', 'education', 'internet_freq']
print("--- Core hypothesis testing results ---")
for var in focus_vars:
    if var in summary.index:
        row = summary.loc[var]
        effect = "positive" if row['Coef'] > 0 else "negative"
        print(f"Variable [{var}]: Effect is {effect} ({row['Coef']:.3f}), statistically {row['Significance']}")
```

# Interpretation of Empirical Results
Based on the regression table and coefficient visualization chart of Model 3, we have drawn the following core sociological findings regarding "The Mechanism of Happiness":
1. Micro verification of the "Easterlin Paradox": The failure of absolute income. This is a highly sociological finding: After controlling for other variables, the impact of absolute income on happiness is not significant ($p > 0.1$, the confidence interval crosses the 0 line). Meanwhile, social class (social_class) shows a significant positive influence ($p < 0.001$). Interpretation: This indicates that in the current social context, "where one feels they stand" is more likely to determine happiness than "how much one actually earns". The theory of relative deprivation (Relative Deprivation) is verified here - people's suffering often does not stem from poverty itself, but from comparisons with others and the low evaluation of one's own status.
2. Health is the capital of revolution: The dominance of health from the coefficient graph. The standardized effect of health (health) is the strongest and highly significant. This once again confirms that physical and mental health is the cornerstone of happiness. Regardless of the social class an individual is in, the deterioration of health will have a "veto" effect on happiness.
3. Technological demystification: The insignificant effect of the Internet. The regression coefficient of Internet usage frequency (internet_freq) is negative and statistically insignificant. This refutes the simple "technology empowerment" argument. Frequent Internet usage does not directly translate into higher life satisfaction; instead, it may bring potential negative emotions due to "information overload" or "social comparison in the online society" (although it does not reach the significant level in this model, the direction of the coefficient is worthy of deep consideration).
4. The hidden benefits of education: Education (education) maintains a marginal significant positive influence in the model. This indicates that education, in addition to indirectly influencing happiness through income transformation, can also directly enhance subjective well-being by improving an individual's cognitive ability and social adaptability.
