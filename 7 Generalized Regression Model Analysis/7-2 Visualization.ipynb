```python
summary_df = pd.DataFrame({
    'coef': model.params,
    'err': model.bse,
    'conf_lower': model.conf_int()[0],
    'conf_upper': model.conf_int()[1]
}).drop('Intercept')

plt.figure(figsize=(10, 8))
plt.errorbar(x=summary_df['coef'], y=summary_df.index, 
             xerr=1.96*summary_df['err'], 
             fmt='o', color='darkblue', ecolor='gray', capsize=3)
plt.axvline(x=0, color='r', linestyle='--', linewidth=1)
plt.title('GLM分析系数图')
plt.xlabel('回归系数 (Coefficient)')
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

This coefficient graph visually presents the direction and relative intensity of the influence of each factor on happiness: health and social class have the strongest positive effect; in terms of education, college, undergraduate, and high school - junior college degrees significantly enhance happiness compared to primary and below levels, while postgraduate and above degrees have no significant impact; in marital status, married status has a positive contribution, while divorce has a significant negative effect; party membership (non-members are the reference) is positively correlated with happiness; internet frequency shows a relatively weak negative impact; and income (income), gender (gender), and type of residence (urban - rural) have insignificant effects. Overall, subjective social factors (such as health, class identity) and stable social identities (such as marriage, political status) have stronger predictive power for happiness, while objective economic factors (such as income) have relatively limited effects.
