```python
def get_model_results(model, model_name):
    """Extract model coefficients and significance stars, formatted as a nice column"""
    # Extract coefficients and p-values
    params = model.params
    pvalues = model.pvalues
    
    # Formatting function: coefficient + significance stars
    def format_func(idx):
        coef = params[idx]
        pval = pvalues[idx]
        stars = ''
        if pval < 0.001: stars = '***'
        elif pval < 0.01: stars = '**'
        elif pval < 0.05: stars = '*'
        return f"{coef:.3f}{stars}"
    
    return pd.Series([format_func(i) for i in params.index], index=params.index, name=model_name)

# Generate columns for the three models
col1 = get_model_results(model1, "Model 1 (Baseline)")
col2 = get_model_results(model2, "Model 2 (Core)")
col3 = get_model_results(model3, "Model 3 (Interaction)")

# Combine into a single table
table = pd.concat([col1, col2, col3], axis=1)

# Add R-squared and sample size N at the bottom
table.loc['R-squared'] = [f"{model1.rsquared:.3f}", f"{model2.rsquared:.3f}", f"{model3.rsquared:.3f}"]
table.loc['Observations'] = [int(model1.nobs), int(model2.nobs), int(model3.nobs)]

# Fill empty values (some variables are not present in Model 1)
table = table.fillna('')

print("\n--- Regression analysis results summary table ---")
print(table)
```
