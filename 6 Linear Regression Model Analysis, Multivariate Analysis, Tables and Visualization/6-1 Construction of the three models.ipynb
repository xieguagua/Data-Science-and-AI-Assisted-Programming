Build three models of hierarchical regression: the baseline model (an empty model), the core model, and the interaction effect model.

```python
# Model 1: Baseline model (demographic control variables)
# happiness ~ gender + marital_status + politics + residence + age
formula1 = 'happiness ~ gender + marital_status + politics + residence + age'
model1 = smf.ols(formula1, data=df).fit()

# Model 2: Core model (adding SES and internet use)
# happiness ~ (variables in Model 1) + income + health + education + social class + internet
formula2 = 'happiness ~ gender + marital_status + politics + residence + age + \
            income + health + education + social_class + internet_freq'
model2 = smf.ols(formula2, data=df).fit()

# Model 3: Interaction effect model (exploring income * health)
# To see if the impact of "income" on happiness varies depending on "health" status?
formula3 = 'happiness ~ gender + marital_status + politics + residence + age + \
            education + social_class + internet_freq + income * health'
model3 = smf.ols(formula3, data=df).fit()

print("All three models have been run!")
print(f"Model 1 (Baseline) R²: {model1.rsquared:.3f}")
print(f"Model 2 (Core) R²: {model2.rsquared:.3f} (Improvement after adding core variables)")
print(f"Model 3 (Interaction) R²: {model3.rsquared:.3f}")
```
