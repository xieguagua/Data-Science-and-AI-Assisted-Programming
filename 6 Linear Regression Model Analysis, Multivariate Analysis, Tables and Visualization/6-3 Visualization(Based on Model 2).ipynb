```python
# Extract data from Model 2 (remove Intercept term)
plot_data = pd.DataFrame({
    'coef': model2.params.drop('Intercept'),
    'pvalue': model2.pvalues.drop('Intercept'),
    'conf_lower': model2.conf_int()[0].drop('Intercept'),
    'conf_upper': model2.conf_int()[1].drop('Intercept')
})

# Sort by coefficient magnitude for better visualization
plot_data = plot_data.sort_values('coef')

# Set colors: significant (P<0.05) in red, non-significant in gray
plot_data['color'] = plot_data['pvalue'].apply(lambda x: '#D62728' if x < 0.05 else '#7F7F7F')

# Plot
plt.figure(figsize=(12, 8))

# Draw horizontal bar chart (with error bars/confidence intervals)
# Error bar length = (upper bound - lower bound) / 2
errors = (plot_data['conf_upper'] - plot_data['conf_lower']) / 2
plt.barh(plot_data.index, plot_data['coef'], xerr=errors, color=plot_data['color'], capsize=4, alpha=0.8)

# Add decorations
plt.axvline(x=0, color='black', linestyle='--', linewidth=1) # Zero reference line
plt.title('Analysis of factors influencing happiness (OLS regression coefficients)', fontsize=16, fontweight='bold')
plt.xlabel('Regression coefficient \n (Positive values indicate increased happiness, negative values indicate decreased happiness)', fontsize=12)
plt.ylabel('Independent variables', fontsize=12)
plt.grid(axis='x', linestyle='--', alpha=0.5)

# Add legend
from matplotlib.patches import Patch
legend_elements = [Patch(facecolor='#D62728', label='Significant (P < 0.05)'),
                   Patch(facecolor='#7F7F7F', label='Not significant (P >= 0.05)')]
plt.legend(handles=legend_elements, loc='lower right')

plt.tight_layout()
plt.show()
```
